```python
### DATA CLEANING
```

```python
df_labels = df["Survived"].copy()
```

```python
titanic = df.drop("Survived", axis=1)
```

```python
titanic.head()
```

```python
titanic.index
```

```python
titanic.info()
```

```python
titanic.drop('Cabin', axis=1, inplace=True)
```

```python
titanic.info()
```

```python
titanic.fillna(value={'Embarked': 'S'}, inplace=True)
```

```python
titanic.drop('Name',axis=1, inplace=True)
```

```python
titanic.drop('Ticket', axis=1, inplace=True)
```

```python
titanic_cat = titanic[["Embarked", "Sex"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
titanic_cat_encoded = ordinal_encoder.fit_transform(titanic_cat)
```

```python
titanic_cat_encoded[:10]
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()
titanic_cat_encoded = cat_encoder.fit_transform(titanic_cat)
titanic_cat_encoded
```

```python
titanic_cat_encoded.toarray()
```

```python
cat_encoder.categories_
```

```python
titanic.info()
```

```python
titanic.drop("PassengerId", axis=1, inplace=True)
titanic_num = titanic.drop(["Sex", "Embarked"], axis=1)
```

```python
titanic_num.info()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
imputer.fit(titanic_num)
```

```python
imputer.statistics_
```

```python
titanic_num.median().values
```

```python
X = imputer.transform(titanic_num)
```

```python
titanic_tr = pd.DataFrame(X, columns=titanic_num.columns, index=titanic_num.index)
```

```python
titanic_tr.info()
```

```python
titanic_tr.describe()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('std_scaler', StandardScaler()),
])

titanic_num_tr = num_pipeline.fit_transform(titanic_num)
```

```python
type(titanic_num_tr)
```

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(titanic_num)
cat_attribs = list(titanic_cat)

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])

titanic_prepared = full_pipeline.fit_transform(titanic)
```

```python
### TRAINING
```

```python
y_train = df_labels
```

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(titanic_prepared, y_train)
```

```python
### EVALUATION
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, titanic_prepared, y_train, cv=2, scoring='accuracy')
```

```python
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(sgd_clf, titanic_prepared, y_train, cv=2)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train, y_train_pred)
```

```python
confusion_matrix(y_train, y_train)
```

```python
from sklearn.metrics import precision_score, recall_score
```

```python
precision_score(y_train, y_train_pred)
```

```python
recall_score(y_train, y_train_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_train, y_train_pred)
```

```python
y_scores = cross_val_predict(sgd_clf, titanic_prepared, y_train, cv=2, method="decision_function")
```

```python
from sklearn.metrics import precision_recall_curve
precisions, recalls, thresholds = precision_recall_curve(y_train, y_scores)
```

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision")
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall")
```

```python
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
plt.show()
```

```python
plt.plot(recalls, precisions)
plt.show()
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_train, y_scores)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--') # Dashed diagonal
```

```python
plot_roc_curve(fpr, tpr)
plt.show()
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_train, y_scores)
```
